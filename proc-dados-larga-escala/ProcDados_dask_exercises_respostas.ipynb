## Resposta Atividade 1

Qual é o anúncio mais popular, ou seja, com mais eventos independente do action?

**Resposta:**  
adId_06

```python
# 1. Agrupamento por 'adId';
# 2. Contagem total de eventos (timestamp) para cada 'adId';
# 3. Encontrar o anúncio mais popular usando nlargest(1).

df.groupby('adId')\
    ['timestamp']\
    .count()\
    .rename('event_count')\
    .nlargest(1)\
    .compute()
```

## Resposta Atividade 2

Quantos clicks gerou a campanha mais popular?

**Resposta:**

```python
# 1. Agrupamento por 'campaignId';
# 2. Contagem total de eventos (timestamp) para cada 'campaignId';
# 3. Encontrar a campanha mais popular usando nlargest(1);
# 4. '.index[0]' para obter a campanha e não a contagem.

most_popular_campaign = df.groupby('campaignId')\
    ['timestamp']\
    .count()\
    .rename('event_count')\
    .nlargest(1)\
    .compute()\
    .index[0]

most_popular_campaign
```

```python
# 1. Filtrando 'df' por 'campaignId' = 'most_popular_campaign' e 'action' = 'click';
# 2. Contagem total de eventos 'click' (timestamp) para a campanha mais popular;

df[(df['campaignId'] == most_popular_campaign) & (df['action'] == 'click')]\
  ['timestamp'].count().compute()
```

## Resposta Atividade 3

Algum usuário só visualizou? Quantas actions foram enviadas por usuário que só visualizou?

**Resposta:**  
Sim, o usuário d99871cb-98b7-4ac5-97a5-b9a26c0f897b enviou apenas 1 action

1. groupby('user_id')['action'].value_counts().rename('events_count'):

O DataFrame é agrupado por 'user_id' e 'action', e em seguida, a função value_counts() é aplicada para calcular a contagem de ocorrências de cada valor único de 'action' para cada usuário. Isso cria uma série Dask com o índice contendo níveis hierárquicos 'user_id' e 'action' e os valores representando as contagens de eventos. Essa série é renomeada para 'events_count';

2. .reset_index():

O índice hierárquico é redefinido para um índice padrão, transformando a série em um DataFrame Dask. Agora, temos as colunas 'user_id', 'action', e 'events_count';

3. .categorize(columns=['action']):

As colunas 'action' são categorizadas, o que pode melhorar o desempenho em certas operações e economizar espaço em memória;

4. .pivot_table(index='user_id', columns='action', values='events_count', aggfunc='sum'):

Uma tabela dinâmica é criada usando as colunas 'user_id' e 'action'. O resultado é um DataFrame onde cada linha representa um usuário, cada coluna representa um tipo de ação ('action'), e os valores são as contagens totais de eventos para cada combinação usuário-ação. O argumento aggfunc='sum' indica que as contagens devem ser somadas em caso de duplicatas.

5. .persist():

O resultado é persistido na memória para otimizar o desempenho em operações subsequentes, tornando a computação mais eficiente.

```python
df.head(3)
```

```python
result_df = df.groupby('user_id')['action']\
    .value_counts()\
    .rename('events_count')\
    .reset_index()\
    .categorize(columns=['action'])\
    .pivot_table(index='user_id', columns='action', values='events_count', aggfunc='sum')\
    .persist()
result_df.compute()
```

```python
result_df[result_df['click'] == 0].compute()
```

```python
# Sim, o usuário d99871cb-98b7-4ac5-97a5-b9a26c0f897b enviou apenas 1 action
```

```python
# '.dt.weekday' obtem o dia da semana referente a cada valor de 'timestamp'
df['week_day'] = df['timestamp'].dt.weekday
df.head(3)
```

```python
# 1: filtra os dados apenas onde a coluna 'action' = 'click';
# 2: agrupa por dia da semana (week_day);
# 3: conta a quantidade de ocorrências pela coluna 'timestamp'.

df[df['action'] == 'click']\
    .groupby('week_day')\
    ['timestamp']\
    .count()\
    .nlargest(7)\
    .compute()
```

## Resposta Atividade 6

Algum anúncio teve um crescimento de clicks ao longo das 3 campanhas?

**Resposta:**  
Sim, o anúncio adId_07

```python
df.head(3)
```

```python
# cria uma coluna nova com os valores iguais a 1
df['event_count'] = 1
df.head(3)
```

'.pivot_table()' cria uma tabela dinâmica usando 'adId' como índice (9 linhas), 'campaignId' como colunas (3 colunas) e 'event_count' como valores. A função de agregação (aggfunc) é definida como 'sum', o que significa que as contagens de eventos serão somadas para cada combinação de 'adId' e 'campaignId'. O resultado é um DataFrame onde cada linha representa um anúncio ('adId'), cada coluna representa uma campanha ('campaignId'), e os valores representam a contagem total de eventos de clique para cada combinação.

```python
ads_df = df[df['action'] == "click"]\
    .categorize(columns=['campaignId'])\
    .pivot_table(index='adId', columns='campaignId', values='event_count', aggfunc='sum')\
    .compute()
ads_df
```

cria uma condição booleana onde cada linha do DataFrame é selecionada se a contagem de eventos para 'campaignId_01' for menor que a contagem de eventos para 'campaignId_02' e, ao mesmo tempo, a contagem de eventos para 'campaignId_02' for menor que a contagem de eventos para 'campaignId_03'. Essa condição é usada para filtrar as linhas do DataFrame.

```python
df_result = ads_df[
    (ads_df['campaignId_01'] < ads_df['campaignId_02']) &\
    (ads_df['campaignId_02'] < ads_df['campaignId_03'])
]
```

 A mensagem inclui o índice (anúncio) da primeira linha (índice 0) do DataFrame resultante (df_result).

```python
print(f"Sim, o anúncio {df_result.index[0]}")
```
