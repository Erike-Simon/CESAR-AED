```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
base16 = pd.read_csv('reclamacoes-fundamentadas-sindec-2016.csv', delimiter =',')
base16
```

```python
base161 = base16.dropna()
```

```python
base161.columns
```

```python
base161.info()
# Regiao, UF, DataAbertura
```

```python
base161.head()
```

```python
atendida = base161[base161['Atendida'] == 'S'].groupby('Regiao').size()
```

```python
atendida.index
```

```python
n_atendida = base161[base161['Atendida'] == 'N'].groupby('Regiao').size()
```

# Levantamento de proporção de reclamações atendidas e não atendidas por região.

```python
# Gráfico

# Eixo x
x = np.arange(len(atendida.index))
largura = 0.3 # Largura das barras

# Criando o gráfico de barras
fig, ax = plt.subplots(figsize=(14,8))

# Plotando as barras
g1 = ax.bar(x - largura/2, atendida.values, largura, label = 'Nº de reclamações atendidas', color = 'green')
g2 = ax.bar(x + largura/2, n_atendida.values, largura, label = 'Nº de reclamações não atendidas', color = 'r')

ax.set_title('RECLAMAÇÕES POR REGIÃO 2016')
ax.set_ylabel('RECLAMAÇÕES')
ax.set_xlabel('REGIÕES')
ax.set_xticks(x)
ax.set_xticklabels(atendida.index)
plt.legend()

# Exibindo o gráfico
plt.show()
```

```python
base161.columns
```

```python
NE = base161[base161['Regiao'] == 'Nordeste'].groupby('UF').size()
```

```python
NE_atendida = base161[(base161['Regiao'] == 'Nordeste') & (base161['Atendida'] == 'S')].groupby('UF').size().sort_index(ascending=True)
NE_atendida
```

```python
NE_n_atendida = base161[(base161['Regiao'] == 'Nordeste') & (base161['Atendida'] == 'N')].groupby('UF').size()
NE_n_atendida['AL'] = 0
NE_n_atendida = NE_n_atendida.sort_index(ascending=True)

```

```python
NE_n_atendida
```

# Levantamento de proporção de reclamações atendidas e não atendidas por estado do nordeste.

A quantidade de reclamações é proporcional a quantidade de população aos estados do nordeste.

```python
# Eixo x
x1 = np.arange(len(NE_atendida.index))
largura = 0.2 # Largura das barras

# Criando o gráfico de barras
fig, ax = plt.subplots(figsize=(14,8))

# Plotando as barras
g3 = ax.bar(x1 - largura/2, NE_atendida.values, largura, label = 'Nº de reclamações atendidas', color = 'green')
g4 = ax.bar(x1 + largura/2, NE_n_atendida.values, largura, label = 'Nº de reclamações não atendidas', color = 'r')

ax.set_title('RECLAMAÇÕES POR ESTADOS DO NORDESTE 2016')
ax.set_ylabel('RECLAMAÇÕES')
ax.set_xlabel('ESTADOS')
ax.set_xticks(x1)
ax.set_xticklabels(NE_atendida.index, rotation=45)
plt.legend()

# Exibindo o gráfico
plt.show()
```
