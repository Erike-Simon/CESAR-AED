# **Tokenização de texto**

````python
from nltk.tokenize import RegexpTokenizer

# O objeto 'RegexpTokenizer' é criado
tokenizer = RegexpTokenizer(r'\w+') # Esta expressão regular corresponde a qualquer palavra no texto.
                                    # O padrão \w+ significa "um ou mais caracteres alfanuméricos".
                                    # O tokenizador irá considerar palavras como unidades.

text = "Eu gosto de café sem açúcar, mas se o café estiver forte eu ponho açúcar. Em nossa casa as pessoas costumão tomar café com bolachas"

# O método 'tokenize' do objeto 'tokenizer' é utilizado para dividir o texto em palavras.
[s.lower() for s in tokenizer.tokenize(text)]
````

# **Remoção de StopWords**

-É necessário ter uma lista de stopwords, no NLTK basta baixar essa lista:
```nltk.download('stopwords')```


````python
nltk.download('stopwords')
````

````python
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords

tokenizer = RegexpTokenizer(r'\w+')
tokens_v0 = [s.lower() for s in tokenizer.tokenize(text)]

stop_words = stopwords.words('portuguese') # fornece uma lista de stopwords em português.
tokens_v1 = [token for token in tokens_v0 if not token in stop_words] # A 'list comprehension' cria uma nova lista, 'tokens_v1',
                                                                      # que contém apenas as palavras que não estão na lista de stopwords.
print(tokens_v0)
print(tokens_v1)
````

Observando a lista de stop words

````python
stop_words.remove("sem") # removendo palavras que ache relevante da lista de stop words
print(stop_words)
````

# **Lemmatization**

````python
nltk.download('wordnet') # O WordNet é um banco de dados lexical da língua inglesa.
from nltk.test.portuguese_en_fixt import setup_module
````

````python
# 'WordNetLemmatizer' é uma classe do NLTK que realiza a lematização, ou seja, reduz palavras à sua forma base.
from nltk.stem import WordNetLemmatizer

wordnet_lemmatizer = WordNetLemmatizer() # Um objeto 'WordNetLemmatizer' é criado.
lemma_words = [wordnet_lemmatizer.lemmatize(word) for word in tokens_v0]

print(tokens_v0)
print(lemma_words)
````

# **Stemming**

````python
from nltk.stem.porter import PorterStemmer

porter_stemmer = PorterStemmer()
stem_words = [porter_stemmer.stem(word) for word in tokens_v0]

print(tokens_v0)
print(stem_words)
````

````python
import pandas as pd
import numpy as np

df = pd.DataFrame(data=[np.zeros(11)], columns=list(set(tokens_v1)), index=["N","TF"])
quant_token = len(list(set(tokens_v1)))

for word in tokens_v0:
  if word in df.keys():
    df[word]["N"] += 1

for token in df.keys():
  df[token]["TF"] = df[token]["N"]/quant_token

df
````

O DataFrame resultante (df) terá duas linhas ("N" e "TF") e colunas correspondentes a palavras únicas. Os valores em "N" representam a contagem de ocorrências de cada palavra, e os valores em "TF" representam a frequência relativa de cada palavra em relação ao número total de tokens únicos.

Quebrando o texto em sentenças.

````python
sentence_tokenizer = nltk.data.load('tokenizers/punkt/portuguese.pickle')
result = sentence_tokenizer.tokenize(text)
print(len(result))

sentences = []
for n, sentence in enumerate(result):
  sentences.append(sentence)
  print(f"{n}: {sentence}")
````

## Pra cada sentença faça:


*   Tokenização
*   Remoção de stopwords usando a lista da lib NLTK (reflita se tem alguma palavra que você colocaria/removeria da lista de stop words, qual seria?)

Tokenização e remoção de stopwords.

````python
tokenizer = RegexpTokenizer(r'\w+')
stop_words = stopwords.words('portuguese')
stop_words.remove("são") # remove 'são' da lista de stop words em português

tokens_cleaned = []
for sentence in sentences:
  token_sentence = tokenizer.tokenize(sentence.lower())
  tokens_v1 = [token for token in token_sentence if not token in stop_words] # lista sem stop words
  tokens_cleaned.append(tokens_v1) # lista que armazena as listas de palavras tokenizadas

  print(token_sentence)
  print(tokens_v1)
  print()
````

## Por fim, compare a Lemmatization com Stemming de cada frase.

````python
for words in tokens_cleaned:
  lemma_words = [wordnet_lemmatizer.lemmatize(word) for word in words]
  stem_words = [porter_stemmer.stem(word) for word in words]

  print(words)
  print(lemma_words)
  print(stem_words)
  print()
````

## Extração de características

````python
dfs = []
for words in tokens_cleaned:
  list_size = len(list(set(words)))
  df = pd.DataFrame(data=[np.zeros(list_size)], columns=list(set(words)), index=["N","TF"])
  quant_token = len(list(set(words)))

  for word in words:
    if word in df.keys():
      df[word]["N"] += 1

  for token in df.keys():
    df[token]["TF"] = df[token]["N"]/quant_token

  dfs.append(df)
````
