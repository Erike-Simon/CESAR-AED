# Transformação dos dados (categoricos/binarização)

## Categorização

Categorização de valores ordinais

Quando categorias apresentarem ordinalidade, utilizamos **LabelEncoder**

```python
#  Importa o módulo de pré-processamento do scikit-learn, que contém várias funções úteis para transformar e preparar dados
from sklearn import preprocessing
```

```python
# Criando uma instância do LabelEncoder, que é uma classe que nos ajuda a codificar rótulos categóricos em valores numéricos
le = preprocessing.LabelEncoder()
```

```python
# Ajusta o codificador aos valores da coluna 'Sex'. Isso permite que o codificador saiba quais rótulos estão presentes nos dados e os associe a valores numéricos específicos
le.fit(df['Sex'])
```

```python
# Transformando os rótulos da cluna 'Sex' em valores numéricos usando o codificador e atribuindo a uma nova coluna 'le_sex'
df['Sex_le'] = le.transform(df['Sex'])
```

```python
df
```

## Binarização

Categorização de valores nominais

Quando valores categóricos **não** apresentam relação de distância entre suas categorias

```python
df['Embarked'].unique()
```

```python
# 'Dropa' os valores 'NaN' da coluna específica 'Embarked'
df.dropna(subset=['Embarked'], inplace=True)
```

```python
# Criar as colunas 'Embarked_q' e 'Embarked_s' usando get_dummies
embarked_dummies = pd.get_dummies(df['Embarked'], prefix='Embarked', drop_first=True)
```

```python
# Concatenar as colunas dummy ao DataFrame df_embarked
df_embarked = pd.concat([df, embarked_dummies], axis=1)
```

```python
df_embarked
```

```python
df_embarked.info()
```

```python
df_embarked.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin', 'Embarked', 'Sex'], inplace=True)
```

# Separação dos dados, holdout

Use o método train_test_split para dividir seus dados em um conjunto treino e teste.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_embarked.drop(columns=['Survived']),
                                                    df_embarked['Survived'],
                                                    test_size=0.3,
                                                    random_state=111)

# random_state=111: Isso define a semente para o gerador de números aleatórios.
# Usar um valor específico para random_state garante que a divisão dos dados seja replicável,
# ou seja, sempre resultará na mesma divisão quando o código for executado novamente com o mesmo valor.
```

```python
pd.DataFrame(X_train)
```

# Normalização dos dados

Hora de para padronizar as variáveis.

Import StandardScaler do Scikit-learn.

```python
X_train.describe()
```

Observa-se que as médias e desvios padrões das colunas (features) 'Age' e 'Fare' estão muito discrepantes dos valores das demais colunas. Isso é um indicativo de que estas necessitam de uma normalização

```python
from sklearn.preprocessing import StandardScaler
```

Converta as variáveis independentes com o StandardScaler

```python
# Um objeto StandardScaler é criado e ajustado aos dados de treinamento das colunas 'Age' e 'Fare'
sc = StandardScaler().fit(X_train[['Age', 'Fare']])
```

Use o método .transform () para transformar os parâmetros em uma versão padronizada.

```python
# Função transform é aplicada aos dados de treinamento e test das colunas 'Age' e 'Fare' usando o scaler previamente ajustado.
# Isso dimensiona (escala) esses recursos com base na média e no desvio padrão calculados durante o ajuste
X_train_scaled = sc.transform(X_train[['Age', 'Fare']])
X_test_scaled = sc.transform(X_test[['Age', 'Fare']])
```

```python
# Dropa as colunas 'Age' e 'Fare' originais do conjunto de dados de treinamento já que não serão utilizadas
X_train.drop(columns=['Age', 'Fare'], inplace=True)
```
